```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/LakshmireddyP/DataSets/main/spam.csv',encoding="ISO-8859-1")
```

```python
df.head()
```

```python
df=df.rename(columns={'v1':'Category','v2':'Messages'})
```

```python
df=df.iloc[:,:2]
df
```

```python
df['spam']=df['Category'].apply(lambda x:1 if x=='spam' else 0)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.Messages,df.spam,test_size=0.25)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python

from sklearn.pipeline import Pipeline
pipe = Pipeline([('Vectorizer',CountVectorizer()),('model',MultinomialNB())])
```

```python
model=pipe.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(y_test,y_pred)
```

```python
report = classification_report(y_test, y_pred)
print(report)
```
