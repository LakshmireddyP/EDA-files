Importing the dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
#Loading data into pandas dataframe
data=pd.read_csv('/content/creditcard.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.tail()
```

```python
#Dataset information
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
#checking the no.of fradulent and legit transactions
data['Class'].value_counts()
```

This data is highly imbalanced
  0---> Normal tranctions
  1---> Fradulent transactions

```python
legit=data[data.Class==0]
fraud=data[data.Class==1]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
#statistical methods of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions
data.groupby('Class').mean()
```

Under Sampling

No of fraud transactions = 153

```python
legit_sample=legit.sample(n=180)
```

```python
legit_sample
```

Concatening two datasets

```python
final_data=pd.concat([legit_sample,fraud],axis=0)
```

```python
final_data.head()
```

```python
final_data.shape
```

```python
final_data.groupby('Class').mean()
```

```python
#splitting data into features and targets
x=final_data.drop(columns='Class',axis=1)
y=final_data['Class']
```

```python
x.head(2)
```

```python
y.head(3)
```

Splitting data into Training data  and testing data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
x_train.shape
```

Model Training

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

Model Evaluation

```python
#Accuracy score of train data
x_pred=model.predict(x_train)
training_score=accuracy_score(x_pred,y_train)
```

```python
print(f'training data accuracy score: {training_score}')
```

```python
#Accuracy score of test data
y_pred=model.predict(x_test)
testing_score=accuracy_score(y_pred,y_test)
```

```python
print(f'testing data accuracy score: {testing_score}')
```
